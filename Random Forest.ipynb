```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
#one-hot method words 
```

```python
dataset=dataset.astype(int)
#to number 0 and 1 state from 24 update
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

# Random Forest

```python
#import RF
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 50, random_state = 0)  #  n_estimators less number decresed accuracy    
regressor.fit(X_train, y_train)
#model creation
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
#to check accuracy
```

```python
#save the model (using pickle)
import pickle
filename="finalized_model_Forest.sav"
pickle.dump(regressor,open(filename,'wb'))
#dump to autosave the .sav file
```

```python
loaded_model=pickle.load(open("finalized_model_Forest.sav",'rb'))
result=loaded_model.predict([[1234,345,4565,1,0]])
#multi inputs given in predict
```
