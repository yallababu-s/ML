```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('50_Startups.csv')
```

```python
#actual columnnames in the database
```

```python
#convert state categorical(word) data to categorical-nominal-onehot-encoding
#removing repeting info first column
dataset=pd.get_dummies(dataset,drop_first=True)
#why drop_first if remain two sates are zero, then its autocallically first state was has value so its unnnecessary
```

```python
#after convert check the columnnames
dataset.columns
```

dataset

```python
#multi inputs and one output declaration
indep=dataset[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']] #x-5inputput
dep=dataset[['Profit']]  # y= one output
```

```python
indep
```

```python
dep
```

```python
#split into training set and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 0.30, random_state = 0)
```

```python
y_test.shape
# 49 rows of profit *30% is 14.7
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)#y=W*x1+b0 for this equation we got value for b1 and bo
```

```python
# Viewing the b1 and bo value  ?
#checking weight and bias 
weight=regressor.coef_
print("Weight of the model={}".format(weight))
bais=regressor.intercept_
print("Intercept of the model={}".format(bais))
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
#y_pred predected output
```

```python
#https://scikit-learn.org/stable/modules/model_evaluation.html
```

```python
#evaluation metric
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```
