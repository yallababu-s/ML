```python
# Reading the Dataset
#dataset = pd.read_csv('/content/drive/MyDrive/50_Startups.csv')
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)
#repetation info sjould be drop
```

```python
datasets
```

```python
datasets.columns
```

```python
indep=datasets[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dep=datasets[['Profit']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 0.30, random_state = 0)
```

```python
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
```

```python
# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid']
}
# Initialize SVR model
svr = SVR()

# Set up GridSearchCV
grid_search = GridSearchCV(svr, param_grid, scoring='r2', cv=5, n_jobs=-1)
#grid_search = GridSearchCV(svr, param_grid, verbose=3 (folds),n_jobs=-1)
#refit=true (best) if false its take last column, verbose=3,n_jobs=1
# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters and R² score on test data
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test) #grid_prediction
r2_test = r2_score(y_test, y_pred)

print("Best Parameters:", grid_search.best_params_) #input parameters=param_grid
print("Best R² score on training set:", grid_search.best_score_)
print("R² score on test set:", r2_test)
```

```python
table=pd.DataFrame.from_dict(grid_search.cv_results_)
print(table)
#AttributeError: module 'pandas' has no attribute 'dataframe' its DataFrame
```

```python
#save the best model
import pickle
filename="final_model_svm_grid.sav"
#pickle.dump(regressor,open(filename,'wb'))
#NameError: name 'regressor' is not defined  regressor=grid_search.fit(X_train, y_train)
pickle.dump(grid_search.fit(X_train, y_train),open(filename,'wb'))
```
